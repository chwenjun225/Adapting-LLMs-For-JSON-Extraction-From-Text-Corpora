```python
df = pd.read_excel("./datasets/top_5_insurance_brands_june_news_and_twitter_only_concat.xlsx")
df.insert(1, "HTML source", "" * len(df))
df.head(5)
```

```python
import chardet
import time 
import requests
from urllib.error import HTTPError
from urllib.error import URLError
from bs4 import BeautifulSoup


# TODO: Tìm hiểu về `USER_AGENTS` và `PROXIES` để có thể thay đổi thông tin user agent và proxy khi cần thiết

# USER_AGENTS = [
#     'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3',
#     'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:54.0) Gecko/20100101 Firefox/54.0',
#     'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/602.3.12 (KHTML, like Gecko) Version/10.1.2 Safari/602.3.12',
# ]

# PROXIES = [
#     'http://proxy1.com:8000',
#     'http://proxy2.com:8031',
# ]

def get_html_page_source(url):    
    try: 
        time.sleep(3)
        resp_html = requests.get(url, timeout=10)
        resp_html.raise_for_status()
    except (HTTPError, URLError, requests.exceptions.RequestException, requests.exceptions.Timeout) as e: 
        print(f"Request failed: {e}")
        return None 
    try: 
        bs = BeautifulSoup(resp_html.content, "html.parser")
        textList = bs.body.find_all("p")
    except AttributeError as e: 
        print(f"Parsing error: {e}")
        return None
    if len(textList) == 0 or not textList: 
        return None
    return bs

i = 6650  
while i < len(df["URL"]):
    url = df["URL"].iloc[i]
    html_page_source = get_html_page_source(url)
    if html_page_source is None:
        print(f"URL at {i} is not found or failed to retrieve...")
    else:
        print(f"URL at {i} found, adding to HTML source...")
        html_page_source = str(html_page_source).replace("\n", " ")
        df.at[df.index[i], "HTML source"] = html_page_source
    i += 1
```

```python
df.to_parquet("./datasets/part0004.parquet", engine="fastparquet")
```
